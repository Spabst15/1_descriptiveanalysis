```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA as sklearnPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.datasets.samples_generator import make_blobs
from pandas.tools.plotting import parallel_coordinates
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
```

```python
anzahlspalten = 3
```

```python
X, y = make_blobs(n_samples=1000, centers=4, n_features= anzahlspalten, cluster_std= 1, random_state=0)
```

```python
data = pd.DataFrame(X)
```

```python
data.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.scatter(data[[0]], data[[1]], data[[2]],c=y)
```

```python
for j in range(0,anzahlspalten):
    for i in range(0,anzahlspalten): 
        if (i>j):
            plt.scatter(X[:,j], X[:,i], c=y)
            plt.show()
```
