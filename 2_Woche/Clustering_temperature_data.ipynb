```python
import numpy as np
import pandas as pd
import scipy, scipy.stats
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
data = pd.read_json("data/monthly_json.json") 
```

```python
pivot_df = data.pivot(index='Date', columns='Source', values='Mean')
```

```python
pivot_df.head()
```

```python
pivot_df.plot(figsize=(20, 8))
```

```python
GCAG = pivot_df[['GCAG']]
```

```python
GCAG.head()
```

```python
GCAG.plot(figsize=(20, 8))
```

```python
GCAG.head()
```

```python
kmeans = KMeans(n_clusters=2).fit(GCAG)
```

```python
GCAG['cluster_nr'] = kmeans.predict(GCAG)
```

```python
GCAG.head()
```

```python
GCAG['mean_1'] = GCAG['GCAG'] * GCAG['cluster_nr']
```

```python
GCAG['mean_0'] = GCAG['GCAG'] * (1 - GCAG['cluster_nr'])
```

```python
GCAG.tail()
```

```python
plt.plot(GCAG[['mean_0', 'mean_1']])
```
