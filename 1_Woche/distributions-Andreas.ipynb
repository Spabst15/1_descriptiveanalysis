# Andere Verteilungen


## Gleichverteilung

```python
# Gleichverteilung
x = scipy.linspace(-1,2,101)
pdf = scipy.stats.uniform.pdf(x)
plt.plot(x, pdf)
```

```python
sample = np.random.uniform(0,1,10000)
fig, ax = plt.subplots()
ax.plot(x, pdf)
ax.hist(sample, 20, density=True, histtype='stepfilled', facecolor='g', alpha=0.75)
plt.show()
```

## Normalverteilung

```python
x = scipy.linspace(-10,10,1000)
pdf = scipy.stats.norm.pdf(x)
plt.plot(x, pdf)
```

```python
sample_size = 1000000
sample = np.random.normal(0,1,sample_size)
fig, ax = plt.subplots()
ax.plot(x, pdf)
ax.hist(sample, 30, density=True, histtype='stepfilled', facecolor='g', alpha=0.75)
plt.show()
print('Median :', np.median(sample))
print('Mean   :', np.mean(sample))
```

```python
mean = np.mean(sample)
#print(sample)
print('Standardabweichung: ', sum(((sample - mean)**2/sample_size))**0.5)
```

```python
df2 = pd.DataFrame(sample)
df2.describe()
```

## Gamma-Verteilung

```python
x = scipy.linspace(0,10,101)
pdf = scipy.stats.gamma.pdf(x,1)
plt.plot(x, pdf)

pdf = scipy.stats.gamma.pdf(x,0.5)
plt.plot(x, pdf)

pdf = scipy.stats.gamma.pdf(x,2)
plt.plot(x, pdf)

pdf = scipy.stats.gamma.pdf(x,5)
plt.plot(x, pdf)

pdf = scipy.stats.gamma.pdf(x,2)
```

```python
sample_size = 100000
sample = np.random.gamma(2,1,sample_size)
fig, ax = plt.subplots()
ax.plot(x, pdf)
ax.hist(sample, 100, density=True, histtype='stepfilled', facecolor='g', alpha=0.75)
plt.show()
```

```python
print('Median :', np.median(sample))
print('Mean   :', np.mean(sample))
```

```python
mean = np.mean(sample)
#print(sample)
print('Standardabweichung: ', sum(((sample - mean)**2/sample_size))**0.5)
```
